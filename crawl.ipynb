```python
def extract_phone_add(soup):
    schools_data = []

    # Find all the div elements with the class 'List_CSN'
    school_divs = soup.find_all('div', class_='List_CSN')

    for div in school_divs:
        # Initialize a dictionary to store the data for each school
        school_info = {}
        
        # The name is within the 'a' element inside the 'div' with class 'text'
        name_tag = div.find('div', class_='text').find('a')
        if name_tag:
            school_info['name'] = name_tag.text.strip()

        # The phone number is within the 'li' element containing the text 'Điện thoại'
        phone_tag = div.find('li', text=lambda text: 'Điện thoại' in text)
        if phone_tag:
            # Clean up the phone number text
            school_info['phone'] = phone_tag.text.replace('Điện thoại:', '').strip()

        # The address is within the 'li' element containing the text 'Địa chỉ'
        address_tag = div.find('li', text=lambda text: 'Địa chỉ' in text)
        if address_tag:
            # Clean up the address text
            school_info['address'] = address_tag.text.replace('Địa chỉ:', '').strip()

        # Add the school_info to our list if it contains both phone and address
        if school_info:
            schools_data.append(school_info)

    return schools_data
```

```python
all_city_names = list(city_dict.values())
city_data = {c: [] for c in all_city_names}
```

```python
session = requests.Session()
session.headers.update(headers)
```

```python
for city in all_city_names:
    initial_response = session.get(url)
    initial_soup = BeautifulSoup(initial_response.text, 'html.parser')
    
    viewstate = initial_soup.find('input', {'id': '__VIEWSTATE'}).get('value')
    viewstate_gen = initial_soup.find('input', {'id': '__VIEWSTATEGENERATOR'}).get('value')
    viewstate_encrypt = initial_soup.find('input', {'id': '__VIEWSTATEENCRYPTED'}).get('value')
    event_target = initial_soup.find('input', {'id': '__EVENTTARGET'}).get('value')
    event_argument = initial_soup.find('input', {'id': '__EVENTARGUMENT'}).get('value')
    last_focus = initial_soup.find('input', {'id': '__LASTFOCUS'}).get('value')
    dnn = initial_soup.find('input', {'id': '__dnnVariable'}).get('value')
    portal_id = initial_soup.find('input', {'id': 'dnn_ctr983_TopMenu_hdfPortalId'}).get('value')
    scroll_top = initial_soup.find('input', {'id': 'ScrollTop'}).get('value')
    
    city_code = city_dict_reverse[city]
    
    form_data = {
        'ScriptManager': 'dnn$ctr736$LabourCenterMap$SearchTask|dnn$ctr736$LabourCenterMap$drdProvince',
        'dnn$ctr1007$SearchHome$txtSearch': '   Từ khóa ...',
        'dnn$ctr983$TopMenu$hdfPortalId': portal_id,
        'dnn$ctr736$LabourCenterMap$txtKey': 'Nhập thông tin tìm kiếm',
        'dnn$ctr736$LabourCenterMap$drdProvince': city_code,
        'dnn$ctr736$LabourCenterMap$drdDistrict': -1,
        'dnn$ctr736$LabourCenterMap$drdVillage': -1,
        '__dnnVariable':dnn,
        'ScrollTop': scroll_top,
        '__LASTFOCUS': last_focus,
        '__VIEWSTATE': viewstate,
        '__VIEWSTATEGENERATOR': viewstate_gen,
        '__VIEWSTATEENCRYPTED': viewstate_encrypt,
        '__EVENTTARGET': 'dnn$ctr736$LabourCenterMap$drdProvince',
        '__EVENTARGUMENT': event_argument,
        '__ASYNCPOST': True
    }
    
    response = session.post(url, data=form_data)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    try:
        n_pages = int(soup.find_all('td', class_='PagingLeft')[0].text.split(' ')[-1].split('/')[-1])
    except:
        n_pages = 1
    
    city_data[city].extend(extract_phone_add(soup))
    
    match = re.search(r'__VIEWSTATE\|([^\|]+)', response.text)
    if match:
        viewstate = match.group(1)
        
    for page in range(2, n_pages+1):
        form_data = {
            'ScriptManager': 'dnn$ctr736$LabourCenterMap$SearchTask|dnn$ctr736$LabourCenterMap$drdProvince',
            'dnn$ctr1007$SearchHome$txtSearch': '   Từ khóa ...',
            'dnn$ctr983$TopMenu$hdfPortalId': portal_id,
            'dnn$ctr736$LabourCenterMap$txtKey': 'Nhập thông tin tìm kiếm',
            'dnn$ctr736$LabourCenterMap$drdProvince': city_code,
            'dnn$ctr736$LabourCenterMap$drdDistrict': -1,
            'dnn$ctr736$LabourCenterMap$drdVillage': -1,
            '__dnnVariable':dnn,
            'ScrollTop': scroll_top,
            '__LASTFOCUS': last_focus,
            '__VIEWSTATE': viewstate,
            '__VIEWSTATEGENERATOR': viewstate_gen,
            '__VIEWSTATEENCRYPTED': viewstate_encrypt,
            '__EVENTTARGET': 'dnn$ctr736$LabourCenterMap$drdProvince',
            '__EVENTARGUMENT': f"Page_{page}",
            '__ASYNCPOST': True
        }
        
        response = session.post(url, data=form_data)
        soup = BeautifulSoup(response.text, 'html.parser')
        city_data[city].extend(extract_phone_add(soup))
    
```
